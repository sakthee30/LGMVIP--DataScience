**Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

**Uploading original image**

```python
img = cv2.imread(path+"dog.jfif")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

**Converting original colour image to grey scale**

```python
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_img)
plt.show()
```

**Inverting the grey scale image**

```python
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

cv2 GaussianBlur() function 

```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

**Inverting the Blurred Image**

```python
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

```python
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
plt.axis('off')
plt.imshow(pencil_img)
plt.show()
```
